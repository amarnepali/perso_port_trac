```python
# lets use yfinance to extract the news update

import yfinance as yf
from finvizfinance.quote import finvizfinance

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

## Create Local LLM Server Connection
from langchain_community.llms import Ollama

# Interactive Web App UI
import streamlit as st

```

```python
llm = Ollama(model="llama3")
```

```python
def classify_sentiment(title):
  output = llm.invoke(f"classify the sentiment as 'POSITIVE'OR 'NEGATIVE' or 'NEUTRAL' with just that one word only, no additional words or reasoning: {title}")
  return output.strip()

def get_news_data(ticker):
  #pull data
  stock = finvizfinance(ticker)
  news_df = stock.ticker_news()

  print(news_df)
```

```python
#scrape broader financial and political news

from GoogleNews import GoogleNews

googlenews = GoogleNews(period='7d', encode='utf-8')
googlenews.search('finance OR politics OR Australia finance')
news_results = googlenews.result()

```

```python
google_news_result_df= pd.DataFrame(news_results)
```

```python
google_news_result_df
```

```python
google_news_data_df = google_news_result_df[['datetime','title','desc','link']].copy()
google_news_data_df
```

```python
# preprocess before putting into LLM
google_news_data_df['title']= google_news_data_df['title'].str.lower()
google_news_data_df['desc']= google_news_data_df['desc'].str.lower()
```

```python
# prompt: import hugging face llm model with pretrained weight for sentiment analysis

!pip install transformers torch

from transformers import pipeline

# Load the sentiment analysis pipeline with a pretrained model
sentiment_analyzer = pipeline("sentiment-analysis")

# Example usage (you can apply this to your dataframes later)
# result = sentiment_analyzer("This is a great movie!")
# print(result)
```

```python
# classify sentiment function applied to each row of google_news_data
google_news_data_df['sentiment'] = google_news_data_df['title'].apply(sentiment_analyzer)
google_news_data_df['sentiment_onDesc'] = google_news_data_df['desc'].apply(sentiment_analyzer)
display(google_news_data_df)
```
